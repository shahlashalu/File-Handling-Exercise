# **1. Write a program that reads a text file and counts the number of lines in the file.**

```python
a = open("stud1.txt", "r")

count = 0
for i in a:
  count += 1
a.close()
print("Number of lines in this file is ",count)
```

# **2. Write a program that reads a text file and prints the first n lines of the file, where n is a user- specified integer.**

```python
n = int(input("Enter required number of lines : "))
a = open("stud1.txt", "r")

count = 0
for i in a:
  if count < n:
    print(i, end = "")
  count += 1

a.close()

```

# **3. Write a program that reads a text file and removes all blank lines from the file.**

```python
x = open("withblanks.txt", "r")

y = open("without blanks.txt", "w")
for i in x:
    if i == "\n":
        continue
    y.write(i)

x.close()
y.close()

print("Check the files 'withblanks.txt' and 'without blanks.txt'")
```

# **4. Write a program that reads a text file and removes all lines that contain a specified string.**

```python
q = open("specific.txt","r")
p = open("without specified.txt","w")
for i in q:
    if i == "Animals\n":
        continue
    p.write(i)

q.close()
p.close()

print("Check the files 'withblanks.txt' and 'without blanks.txt'")
```

# **5. Write a program that reads a text file and removes all punctuation from the text.**

```python
punc = open("punctuation.txt", "r")
wp = open("without punctuations.txt", "w")
for i in punc.read():
    if i.isalnum() or i.isspace():
        wp.write(i)

punc.close()
wp.close()

print("Check the files 'punctuation.txt' and 'without punctuations.txt'")
```

# **6. Write a program that reads a text file and capitalizes the first letter of each word.**

```python
wcap = open("without capitalize.txt", "r")

cap = open("with capitalize.txt","w")

cap.write(wcap.read().title())
wcap.close()
cap.close()

print("Check the files 'without capitalize.txt' and 'with capitalize.txt' ")

```

# **7. Write a program that reads a text file and converts all the text to uppercase.**

```python
ac = open("without capitalize.txt", "r")

c = open("all capital.txt", "w")

c.write(ac.read().upper())

ac.close()
c.close()

print("Check the files 'without capitalize.txt' and 'all capital.txt' ")
```

# **8. Write a program that reads a text file and convens all the text to lowercase.**

# **9. Write a program that reads a text file and replaces all occurrences of one word with another word.**

```python
rpt = open("rep.txt", "r")
rplc = open("rpl.txt", "w")

rplc.write(rpt.read().replace("John","Ebin").replace("Emily","Mia"))

rpt.close()
rplc.close()

print("Check the files 'rep.txt' and 'rpl.txt' ")
```
